```python
%matplotlib inline
from collections import defaultdict
import json

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression


rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['lines.linewidth'] = 2
rcParams['font.size'] = 14
rcParams['font.family'] = 'StixGeneral'
```

```python
model_data = pd.read_csv("./Data/Env_factor_9Var.csv")
model_data.tail()
```

```python
variable_list = ["MSCI","fed_funds","EURO_CCI","SP500","OECD_Maj6","ISM2","trade2","water_supply2","trade_balance"]
```

```python
reg = LinearRegression()
lag = 6
for i in range(24)[14:24]:
    N = 3*i
    x_train = model_data[variable_list][:N]
    y_train = model_data[["Environment"]][lag:N+lag]
    x_test = model_data[variable_list][N:-lag-1]
    y_test = model_data[["Environment"]][N+lag:-1]
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    time = range(73)[N+1:73]
    plt.plot(time,predicted_y)
```

```python
r_square=[]
y_hat=[]
yHat = []
training_error=[]
test_error=[]
reg = LinearRegression()
reg2 = LinearRegression()
lag = 6
p = len(variable_list)
for i in range(73)[1:]:
    x_train = model_data[variable_list][:-i-lag].values
    y_train = model_data[["Environment"]][lag:-i].values
    x_test = model_data[variable_list][-lag-i:-lag-i+1].values
    if i == 1:
        y_test = model_data[["Environment"]][-i:].values
    else: 
        y_test = model_data[["Environment"]][-i:-i+1].values
    x_validation = model_data[variable_list][-i-lag:-lag].values
    y_validation = model_data[["Environment"]][-i:].values
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    y_hat.append(predicted_y)
    yHat.append(reg.predict(x_train[-1]))
    rs = reg.score(x_train,y_train)
    adjusted_rs = 1-(1-rs)*(len(x_train)-1)/(len(x_train)-p-1)
    r_square.append(adjusted_rs)
    y_training_fit = reg.predict(x_train)
    y_test_fit = reg.predict(x_validation)
    training_error.append(np.sqrt(np.mean((y_train-y_training_fit)**2)))
    test_error.append(np.sqrt(np.mean((y_validation-y_test_fit)**2)))
x = model_data[variable_list][:-1-lag].values
y = model_data[["Environment"]][-len(y_hat)-1:-1].values
y_hat = y_hat[::-1] # 37...73
yHat = yHat[::-1]
r_square = r_square[::-1]
training_error = training_error[::-1]
test_error = test_error[::-1]
t = range(73)[-len(y_hat):]
residual = map(lambda x, y: x - y, y, y_hat)
training_size = range(73)[-len(y_hat):]
for x in [y,y_hat,residual]:
    for index, item in enumerate(x):
        x[index] = float(item)
```

```python
plt.plot(t,y,label="y")
plt.plot(t,yHat,label="y_hat")
plt.legend()
```

```python
plt.plot(t,y,label="y")
plt.plot(t,y_hat,label="preidicted_y")
plt.legend()
```

```python
plt.scatter(residual,y_hat)
plt.xlabel("fitted value", size=18)
plt.ylabel("residual",size=18)
plt.title("residual vs. fitted value",size=18)
```

```python
plt.plot(t,r_square)
plt.xlabel("training set size",size=18)
plt.ylabel("adjusted R square", size=18)
plt.title("adjusted R square",size=18)
```

```python
plt.plot(training_size,training_error,label='training error')
plt.plot(training_size,test_error,label='test error')
plt.xlabel("training set size",size=18)
plt.ylabel("error",size=18)
plt.legend()
```
