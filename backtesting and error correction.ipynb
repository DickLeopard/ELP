```python
model_data = pd.read_csv("./Data/model2.csv")
model_data.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
error = []
reg = LinearRegression()
lag = 6
for i in range(24)[12:24]:
    N = 3*i
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:N]
    y_train = model_data[["DHR_Avg"]][lag:N+lag]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][N:-lag-1]
    y_test = model_data[["DHR_Avg"]][N+lag:-1]
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    time = range(73)[N+1:73]
    plt.plot(time,predicted_y)
```

```python
forecast_error=[]
error = []
for t in range(38)[1:]:
    x = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:-lag-t]
    y = model_data[["DHR_Avg"]][lag:-t]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][-lag-t:-lag-t+1].values
    if t == 1:
        y_test = model_data[["DHR_Avg"]][-t:].values
    else: 
        y_test = model_data[["DHR_Avg"]][-t:-t+1].values
    reg.fit(x,y)
    predicted_y = reg.predict(x_test)
    f = y_test - predicted_y
    forecast_error.append(f)
    e = y.values[-1] - reg.predict(x.values[-1])
    error.append(e)
for index, item in enumerate(forecast_error):
    forecast_error[index] = float(item)
for index, item in enumerate(error):
    error[index] = float(item)
```
