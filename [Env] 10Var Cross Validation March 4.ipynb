```python
%matplotlib inline
from collections import defaultdict
import json

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression


rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['lines.linewidth'] = 2
rcParams['font.size'] = 14
rcParams['font.family'] = 'StixGeneral'
```

```python
model_data = pd.read_csv("./Data/Env_factor_10Var.csv")
model_data.tail(5)
```

```python
variable_list = ["MSCI","fed_funds","EURO_CCI","SP500","OECD_Maj6","ISM2","trade2","water_supply","trade_balance","water_supply2"]
```

```python
r_square=[]
training_error=[]
test_error=[]
training_set_size = 54
reg = LinearRegression()
lag = 6
p = len(variable_list)
for i in range(p+1)[1:]:
    x_train = model_data[variable_list[:i]][:training_set_size].values
    y_train = model_data[["Environment"]][lag:training_set_size+lag].values
    x_test = model_data[variable_list[:i]][training_set_size:-lag].values
    y_test = model_data[["Environment"]][training_set_size+lag:].values
    reg.fit(x_train, y_train)
    rs = reg.score(x_train,y_train)
    adjusted_rs = 1-(1-rs)*(len(x_train)-1)/(len(x_train)-p-1)
    r_square.append(adjusted_rs)
    y_training_fit = reg.predict(x_train)
    y_test_fit = reg.predict(x_test)
    training_error.append(np.sqrt(np.mean((y_train-y_training_fit)**2)))
    test_error.append(np.sqrt(np.mean((y_test-y_test_fit)**2)))
model_complexity = range(p+1)[1:]
```

```python
plt.plot(model_complexity,r_square)
plt.xlabel("number of variables",size=18)
plt.ylabel("adjusted R square",size=18)
plt.title("adjusted R square vs. model complexity",size=18)
```

```python
plt.plot(model_complexity,training_error,label="training error")
plt.plot(model_complexity,test_error,label="test error")
plt.xlabel("number of variables",size=18)
plt.ylabel("error",size=18)
plt.title("errors vs. model complexity",size=18)
plt.legend()
```
