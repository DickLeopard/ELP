```python
model_data = pd.read_csv("./Data/environment_regression.csv")
model_data.tail(5)
```

```python
variable_list=["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent","Dollar Index","ISM1","UDS/CNY","ISM2","EURO CCI","unemploy","fx_index_1","fx_index_2","Water2","JPM2","OECD2","trade2","Econ2"]
```

```python
reg = LinearRegression()
lag = 6
for i in range(24)[8:24]:
    N = 3*i
    x_train = model_data[variable_list][:N]
    y_train = model_data[["Environment"]][lag:N+lag]
    x_test = model_data[variable_list][N:-lag-1]
    y_test = model_data[["Environment"]][N+lag:-1]
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    time = range(73)[N+1:73]
    plt.plot(time,predicted_y)
```

```python
r_square=[]
y_hat=[]
training_error=[]
test_error=[]
reg = LinearRegression()
lag = 6
p = len(variable_list)
for i in range(73)[1:]:
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent","Dollar Index","ISM1","UDS/CNY","ISM2","EURO CCI","unemploy","fx_index_1","fx_index_2","Water2","JPM2","OECD2","trade2","Econ2"]][:-i-lag].values
    y_train = model_data[["Environment"]][lag:-i].values
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent","Dollar Index","ISM1","UDS/CNY","ISM2","EURO CCI","unemploy","fx_index_1","fx_index_2","Water2","JPM2","OECD2","trade2","Econ2"]][-lag-i:-lag-i+1].values
    if i == 1:
        y_test = model_data[["Environment"]][-i:].values
    else: 
        y_test = model_data[["Environment"]][-i:-i+1].values
    x_validation = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent","Dollar Index","ISM1","UDS/CNY","ISM2","EURO CCI","unemploy","fx_index_1","fx_index_2","Water2","JPM2","OECD2","trade2","Econ2"]][-i-lag:-lag].values
    y_validation = model_data[["Environment"]][-i:].values
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    y_hat.append(predicted_y)
    rs = reg.score(x_train,y_train)
    adjusted_rs = 1-(1-rs)*(len(x_train)-1)/(len(x_train)-p-1)
    r_square.append(adjusted_rs)
    y_training_fit = reg.predict(x_train)
    y_test_fit = reg.predict(x_validation)
    training_error.append(np.sqrt(np.mean((y_train-y_training_fit)**2)))
    test_error.append(np.sqrt(np.mean((y_validation-y_test_fit)**2)))
y_hat = y_hat[::-1] # 37...73
r_square = r_square[::-1]
training_error = training_error[::-1]
test_error = test_error[::-1]
y = model_data[["Environment"]][-len(y_hat):].values
t = range(74)[-len(y_hat):]
residual = map(lambda x, y: x - y, y, y_hat)
training_size = range(73)[-len(y_hat):]
for x in [y,y_hat,residual]:
    for index, item in enumerate(x):
        x[index] = float(item)
```

```python
plt.plot(t,y,label="y")
plt.plot(t,y_hat,label="y_hat")
plt.legend()
```

```python
plt.scatter(residual,y_hat,label="residual vs. fitted value")
plt.legend()
```

```python
plt.plot(t,r_square, label="adjusted R square")
plt.xlabel("training set size",size=18)
plt.ylabel("adjusted R square", size=18)
plt.legend()
```

```python
plt.plot(training_size,training_error,label='training error')
plt.plot(training_size,test_error,label='test error')
plt.xlabel("training set size",size=18)
plt.ylabel("error",size=18)
plt.legend()
```
