```python
model_data = pd.read_csv("./Data/environment_regression.csv")
model_data.tail(5)
```

```python
r_square=[]
reg = LinearRegression()
lag = 6
p = 5
window_width = w = 36
for i in range(w+2)[1:]:
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent","Dollar Index","ISM1","UDS/CNY","ISM2","EURO CCI","unemploy","fx_index_1","fx_index_2","Water2","JPM2","OECD2","trade2","Econ2"]][i-1:i-1+w].values
    y_train = model_data[["Environment"]][i-1+lag:i-1+w+lag].values
    reg.fit(x_train, y_train)
    rs = reg.score(x_train,y_train)
    adjusted_rs = 1-(1-rs)*(len(x_train)-1)/(len(x_train)-p-1)
    r_square.append(adjusted_rs)
t = range(w+2)[1:]
```

```python
plt.plot(t,r_square,label="adjusted R square")
plt.xlabel("training set start point", size=18)
plt.ylabel("adjusted R square", size=18)
plt.title("shift window adjusted R square, window width=36")
plt.legend()
```
