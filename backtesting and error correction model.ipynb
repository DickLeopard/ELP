```python
model_data = pd.read_csv("./Data/model2.csv")
model_data.tail()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
lag = 6
for i in range(24)[12:24]:
    N = 3*i
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:N]
    y_train = model_data[["DHR_Avg"]][lag:N+lag]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][N:-lag-1]
    y_test = model_data[["DHR_Avg"]][N+lag:-1]
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    time = range(73)[N+1:73]
    plt.plot(time,predicted_y)
    
```

from the graph above, we can see that the model derived from the regression using data from 2008 to 2014 are quite different from the model derived from the regression using data from 2011Q4 to 2014, which indicates the firm's acquiring of health care sector in 2011Q4

Get the error list and forcast_error list, starting point set at 20.
error: N = 20...72, in total 53; forcast_error: N = 20...71, in total 52

```python
error = []
forecast_error=[]
for t in range(54)[1:]:
    x = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:-lag-t]
    #when t=1, x has 72 data
    y = model_data[["DHR_Avg"]][lag:-t]
    #when t=1, y has 72 data
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][-lag-t:-lag-t+1].values
    if t == 1:
        y_test = model_data[["DHR_Avg"]][-t:].values
    else: 
        y_test = model_data[["DHR_Avg"]][-t:-t+1].values
    reg.fit(x,y)
    predicted_y = reg.predict(x_test)
    f = y_test - predicted_y #@t=1: N=72, y73-y73_hat, regression at 72D
    forecast_error.append(f)# remove the F73(N=72) after wards
    e = y.values[-1] - reg.predict(x.values[-1])#@t=1:N=72, y72-y72_hat, regression at 72D
    error.append(e)#e72(N=72)will not be used in regression but will only be used in forcast F73(N=72)

#error and forecast_error are data used in regression
```

```python
Y_hat = []
Y_tuta = []
difference = []
reg1 = LinearRegression()
reg2 = LinearRegression()
for t in range(38)[1:]:
    x = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:-lag-t]
    y = model_data[["DHR_Avg"]][lag:-t]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][-lag-t:-lag-t+1].values
    if t == 1:
        y_test = model_data[["DHR_Avg"]][-t:].values
    else: 
        y_test = model_data[["DHR_Avg"]][-t:-t+1].values
    
    reg1.fit(x,y)
    error_regression = error[:-1]#@t=1, have 52 data from N=20...71
    forecast_error_regression = forecast_error[:-1]
    reg2.fit(error_regression,forecast_error_regression)
    predicted_y = reg1.predict(x_test) # y_hat73
    errorcrrection_y = predicted_y+reg2.predict(error[-t])
    Y_hat.append(predicted_y)
    Y_tuta.append(errorcrrection_y)
    difference.append(reg2.predict(error[-t]))
Y = model_data[["DHR_Avg"]][42:].values#y37...y73
T = range(74)[37:][::-1]
diff1 = map(lambda x, y: x - y, Y, Y_hat)
diff2 = map(lambda x, y: x - y, Y, Y_tuta)
diff3 = map(lambda x, y: x - y, diff1, diff2)
for x in [Y,Y_hat,Y_tuta,difference,diff1,diff2,diff3]:
    for index, item in enumerate(x):
        x[index] = float(item)
```

```python
plt.plot(T,Y,label='Y')
plt.plot(T,Y_hat,label='Y_hat')
plt.plot(T,Y_tuta,label='Y_tuta')
plt.legend()
```

```python
plt.plot(T,diff1,label="Y-Y_hat")
plt.plot(T,diff2,label="Y-Y_tuta")
plt.legend()
```
