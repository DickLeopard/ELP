```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
lag = 6
for i in range(24)[12:24]:
    N = 3*i
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:N]
    y_train = model_data[["DHR_Avg"]][lag:N+lag]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][N:-lag-1]
    y_test = model_data[["DHR_Avg"]][N+lag:-1]
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    time = range(73)[N+1:73]
    plt.plot(time,predicted_y)
    
```

Get the error list and forcast_error list, starting point set at 20.
error: N = 20...72, in total 53; forcast_error: N = 20...71, in total 52

```python
error = []
forecast_error=[]
for t in range(54)[1:]:
    x = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:-lag-t]
    #when t=1, x has 72 data
    y = model_data[["DHR_Avg"]][lag:-t]
    #when t=1, y has 72 data
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][-lag-t:-lag-t+1].values
    if t == 1:
        y_test = model_data[["DHR_Avg"]][-t:].values
    else: 
        y_test = model_data[["DHR_Avg"]][-t:-t+1].values
    reg.fit(x,y)
    predicted_y = reg.predict(x_test) 
    f = y_test - predicted_y #@t=1: N=72, y73-y73_hat, regression at 72D
    forecast_error.append(f)# remove the F73(N=72) after wards
    e = y.values[-1] - reg.predict(x.values[-1])#@t=1:N=72, y72-y72_hat, regression at 72D
    error.append(e)#e72(N=72)will not be used in regression but will only be used in forcast F73(N=72)
error = error[::-1] #error = [e20...e72]
forecast_error = forecast_error[::-1] #forecast_error = [f21...f73]
```
