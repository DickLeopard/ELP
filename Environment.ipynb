```python
r_square=[]
y_hat=[]
reg = LinearRegression()
lag = 6
for i in range(37)[1:]:
    x_train = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][:-i-lag]
    y_train = model_data[["Environment"]][lag:-i]
    x_test = model_data[["Water_Avg","JPM_Global_PMI","OECD_Maj6","trade_balance","Econ_Sent"]][-lag-i:-lag-i+1].values
    if i == 1:
        y_test = model_data[["Environment"]][-i:].values
    else: 
        y_test = model_data[["Environment"]][-i:-i+1].values
    reg.fit(x_train, y_train)
    predicted_y = reg.predict(x_test)
    y_hat.append(predicted_y)
    r_square.append(reg.score(x_train,y_train))
    time = range(73)[N+1:73]
y_hat = y_hat[::-1] # 38...73
r_square = r_square[::-1]
y = model_data[["Environment"]][43:].values
t = range(74)[38:]
residual = map(lambda x, y: x - y, y, y_hat)
for x in [y,y_hat,residual]:
    for index, item in enumerate(x):
        x[index] = float(item)
```

```python
plt.plot(t,y,label="y")
plt.plot(t,y_hat,label="y_hat")
plt.legend()
```
